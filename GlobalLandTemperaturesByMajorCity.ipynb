```python
import pandas as pd
```

```python
temp = pd.read_csv('GlobalLandTemperaturesByMajorCity.csv')
temp
```

```python
temp.columns
```

```python
temp.Country.unique()
```

```python
temp.City.unique()
```

```python
temp['dt'] = pd.to_datetime(temp['dt'])
```

```python
BH = temp.loc[temp['City']=="Belo Horizonte"]
RJ = temp.loc[temp['City']=="Rio De Janeiro"]
Berlin = temp.loc[temp['City']=="Berlin"]
AddisAbeba = temp.loc[temp['City']=="Addis Abeba"]
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(
                x=BH['dt'],
                y=BH['AverageTemperature'],
                name="BH",
                opacity=0.8))
fig.add_trace(go.Scatter(
                x=RJ['dt'],
                y=RJ['AverageTemperature'],
                name="RJ",
                opacity=0.8))
fig.add_trace(go.Scatter(
                x=AddisAbeba['dt'],
                y=AddisAbeba['AverageTemperature'],
                name="AddisAbeba",
                opacity=0.8))
fig.add_trace(go.Scatter(
                x=Berlin['dt'],
                y=Berlin['AverageTemperature'],
                name="Berlin",
                opacity=0.8))
fig.show()
```
